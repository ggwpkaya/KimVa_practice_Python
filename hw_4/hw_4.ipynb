```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/train.csv')

print(data.head())
```

```python
# Визуализация для признаков Survived, Pclass, Age, Sex, Parch
plt.figure(figsize=(12, 8))

# Survived
plt.subplot(2, 3, 1)
data['Survived'].value_counts().plot(kind='bar', color=['skyblue', 'orange'])
plt.title('Распределение Survived')
plt.xlabel('Survived (0=Not survived, 1=Survived)')
plt.ylabel('Количество')

# Pclass
plt.subplot(2, 3, 2)
data['Pclass'].value_counts().plot(kind='bar', color='lightgreen')
plt.title('Распределение Pclass')
plt.xlabel('Pclass')
plt.ylabel('Количество')

# Age
plt.subplot(2, 3, 3)
data['Age'].plot(kind='hist', bins=20, color='purple', edgecolor='black')
plt.title('Распределение Age')
plt.xlabel('Age')
plt.ylabel('Частота')

# Sex
plt.subplot(2, 3, 4)
data['Sex'].value_counts().plot(kind='bar', color=['pink', 'lightblue'])
plt.title('Распределение Sex')
plt.xlabel('Sex')
plt.ylabel('Количество')

# Parch
plt.subplot(2, 3, 5)
data['Parch'].value_counts().plot(kind='bar', color='coral')
plt.title('Распределение Parch')
plt.xlabel('Parch')
plt.ylabel('Количество')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=data, x='Age', color='lightblue')
plt.title('Boxplot для Age')
plt.xlabel('Age')
plt.show()
```

```python
# Pie chart для Survived
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
data['Survived'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['orange', 'skyblue'])
plt.title('Доли Survived')
plt.ylabel('')

# Pie chart для Pclass
plt.subplot(1, 2, 2)
data['Pclass'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['lightgreen', 'lightblue', 'pink'])
plt.title('Доли Pclass')
plt.ylabel('')

plt.tight_layout()
plt.show()
```

```python
#pairplot для всех числовых переменных
sns.pairplot(data.select_dtypes(include=['float64', 'int64']), diag_kind='kde', corner=True)
plt.suptitle('Pairplot для числовых переменных', y=1.02)
plt.show()
```

```python
#sunburst график
fig = px.sunburst(data,
                  path=['Pclass', 'Sex'],
                  values='PassengerId',
                  title='Sunburst: Количество пассажиров по классам и полу')

fig.update_traces(textinfo='label+percent entry')
fig.show()
```
