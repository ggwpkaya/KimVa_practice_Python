```python
# Базовый URL
base_url = "https://lifehacker.ru"

parsed_urls = [
    "/anons-jbl-flip-7/",
    "/microsoft-obnovila-pusk/",
    "/anons-asus-fragrance-mouse/",
    "/nejroseti-dlya-generacii-teksta/",
    "/kak-klonirovat-prilozhenie-android/",
    "/luchshie-igry-na-android/",
    "/sdelat-oblako-bezopasnee/",
    "/anons-redmagic-10-pro-golden-saga-edition/",
    "/10-wireless-headphones/",
    "/anons-oukitel-wp100-titan/"
]

result = []

# Проходим по каждой ссылке с помощью tqdm для отображения прогресса
for url in tqdm(parsed_urls):
    article = {}
    article_url = f'{base_url}{url}'  # Формируем полный URL

    response = requests.get(article_url)
    soup = BeautifulSoup(response.text, 'html.parser')

    article['title'] = soup.find('h1', class_='article-card__title').text
    article['text'] = soup.find('div', class_='single-article').text

    # Добавляем данные статьи в результат
    result.append(article)

# Создаем DataFrame из полученных данных
pd.set_option('display.max_colwidth', 400)

data = pd.DataFrame(result)
data.head()

```
