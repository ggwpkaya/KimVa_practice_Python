```python
import requests
import pandas as pd
```

```python
api_base_url="https://www.anapioficeandfire.com/api"
```

```python
#Работа с API
#Информация обо всех книгах
books_list = requests.get(
    f'{api_base_url}/books').json()
df_books = pd.DataFrame(books_list)
df_books
```

```python
#Информация о домах
houses_info = requests.get(
    f'{api_base_url}/houses').json()
houses_df = pd.DataFrame(houses_info)
df_houses = pd.DataFrame(houses_df)
df_houses
```

```python
#Получить информацию обо всех домах Вестероса, у которых есть девиз
houses_with_words_df = requests.get(
    f'{api_base_url}/houses?hasWords=True').json()

df_houses_ww = pd.DataFrame(houses_with_words_df)
df_houses_ww
```
